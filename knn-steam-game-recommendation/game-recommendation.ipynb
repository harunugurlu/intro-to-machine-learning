```python
# I will use the KNN algorithm on the following columns:
# Windows, Mac, Linux, Positive and Negative (The ratio will give the review score),
# Publishers, Categories, Genres, Tags

# Extracting those series into a new data frame:
features_for_rec = steam_games[['Name', "Positive", "Negative", "Publishers", "Categories", "Genres", "Tags", "Windows", "Mac", "Linux"]]
```

```python
# Trying to figuring out the missing values distribution to see which columns I'll need to work on
print(features_for_rec.isnull().mean())
# Tags column has considerable amount of missing rows (almost 20%)
```

```python
print(features_for_rec.info())
print("")
print(features_for_rec.describe(include=object))
# The game Shadow of the Tomb Raider: Definitive Edition has more than one records.
```

```python
columns_for_rec = list(features_for_rec.columns)

# Removing trailing characters and whitespaces
str_columns = [c for c in columns_for_rec if c not in ('Positive', 'Negative', 'Windows', 'Mac', 'Linux')]

for c in str_columns:
    features_for_rec = features_for_rec.assign(**{c: features_for_rec[c].str.strip()})

# Replacing NaN with empty string
features_for_rec.fillna("")
```

```python
# There are duplicated Names, remove them
features_for_rec = features_for_rec.drop_duplicates('Name')

print(features_for_rec.describe(include=object))
```
